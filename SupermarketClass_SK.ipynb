```python
class Supermarket:
    """manages multiple Customer instances that are currently in the market.
    """

    def __init__(self, hour=7):
        # a list of Customer objects
        self.customers = []
        self.new_cust = []
        self.c_id = 0
        self.minutes = 0
        self.current_time = 0
        self.data = []
        self.timestamps = []
        #self.hour = hour
        if hour < 7:
            self.hour = 7
        elif hour > 21:
            self.hour = 21
        else:
            self.hour = hour
        
        
    def __repr__(self):
        return f'Supermarket("{self.customers}", "{self.current_time}")'
    
    
    def get_time(self):
        """current time in HH:MM format.
        """
        
        minutes = self.minutes
        if self.minutes > 59:
            self.minutes = 0
            self.hour += 1
        self.current_time = f'{str(self.hour).zfill(2)}:{str(minutes).zfill(2)}'
        return self.current_time
    
    
    def timeline(self):
        """add current time in HH:MM format to list in time.
        """
        
        #self.current_time = self.get_time()
        self.get_time()
        self.timestamps.append(self.current_time)
        
        
    def print_customers(self):
        """prints list with all active customers.
        """
        
        return self.customers
        
    
    def move_customers(self):
        """moves existing customers within the supermarket.
        """
        for cust in self.customers:
            cust.next_state()
            print (cust)

            
    def add_new_customers(self):
        """randomly creates new customers.
        """
        
        fake = Faker()
        self.get_time()
        time_gr_1=["7", "9", "10", "20"]
#         time_gr_2=["8", "18"]
#         time_gr_3=["11", "12", "15"]
#         time_gr_4=["13", "14", "16", "17"]
#         time_gr_5=["19"]
#         time_gr_6=["21"]
        if self.current_time[:2] == any(time_gr_1):
            for i in range(13):
                self.c_id += 1
                self.new_cust = Customer(self.c_id, fake.name(), 'entrance', np.random.randint(1,500))  
                self.customers.append(self.new_cust)
                print (self.new_cust)
#         elif self.current_time[:2] == any(time_gr_2):
#             for i in range(18):
#                 self.c_id += 1
#                 self.new_cust = Customer(self.c_id, fake.name(), 'entrance', np.random.randint(1,500))  
#                 self.customers.append(self.new_cust)
#                 print (self.new_cust)
#         elif self.current_time[:2] == any(time_gr_3):
#             for i in range(10):
#                 self.c_id += 1
#                 self.new_cust = Customer(self.c_id, fake.name(), 'entrance', np.random.randint(1,500))  
#                 self.customers.append(self.new_cust)
#                 print (self.new_cust)
#         elif self.current_time[:2] == any(time_gr_4):
#             for i in range(15):
#                 self.c_id += 1
#                 self.new_cust = Customer(self.c_id, fake.name(), 'entrance', np.random.randint(1,500))  
#                 self.customers.append(self.new_cust)
#                 print (self.new_cust)
#         elif self.current_time[:2] == any(time_gr_5):
#             for i in range(20):
#                 self.c_id += 1
#                 self.new_cust = Customer(self.c_id, fake.name(), 'entrance', np.random.randint(1,500))  
#                 self.customers.append(self.new_cust)
#                 print (self.new_cust)
#         elif self.current_time[:2] == any(time_gr_6):
#             for i in range(7):
#                 self.c_id += 1
#                 self.new_cust = Customer(self.c_id, fake.name(), 'entrance', np.random.randint(1,500))  
#                 self.customers.append(self.new_cust)
#                 print (self.new_cust)
        else:
            self.c_id += 1
            self.new_cust = Customer(self.c_id,fake.name(), 'entrance', np.random.randint(1,500))   
            self.customers.append(self.new_cust)
            print (self.new_cust)

        
    def remove_existing_customers(self):
        """removes every customer that is not active any more.
        """
        
        for cust in self.customers:
            if cust.state == 'checkout':
                self.customers.remove(cust) #removes the first matching element (which is passed as an argument) from the list
                print(f'Customer {cust.name} has been churned!')
         
        
    def write_all_cust(self):
        for cust in self.customers:
            self.data.append([cust.c_id, cust.name, cust.state, cust.budget, self.get_time()])

    def simulate(self):
        """propagates all customers to the next state.
        """
        
        self.minutes += 1
        self.move_customers()
        self.add_new_customers()
        self.write_all_cust()
        self.remove_existing_customers()
        self.timeline()
    
```

```python
def marketdata_to_df(market):
    """Given the data from a supermarket (Supermarket() object),
    returns the data as DataFrame.
    """
    
    df = pd.DataFrame.from_records(market.data, columns=['cust_id', 'cust_name', 'cust_state', 'budget', 'timestamp'])
    return df
```
