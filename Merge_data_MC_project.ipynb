```python
# Import general libraries
import pandas as pd
```

## Read data files

```python
df_mon = pd.read_csv('data/monday.csv', sep=";")
```

```python
df_tue = pd.read_csv('data/tuesday.csv', sep=";")
```

```python
df_wed = pd.read_csv('data/wednesday.csv', sep=";")
```

```python
df_thu = pd.read_csv('data/thursday.csv', sep=";")
```

```python
df_fri = pd.read_csv('data/friday.csv', sep=";")
```

## Merge data

```python
df = pd.concat([df_mon, df_tue, df_wed, df_thu, df_fri])
```

```python
df
```

```python
df.shape
```

```python
new_col = range(1,len(df)+1)
df.insert(loc=2, column='new_id', value=new_col)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['weekday'] = df['timestamp'].dt.day_name()
```

```python
df['time'] = df['timestamp'].dt.time
```

```python
df.set_index('timestamp')
```

## Save data

```python
df.to_csv('data/data.csv', sep=";")
```
